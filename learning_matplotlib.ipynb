```python
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'key1':['a','a','b','b','a'],
    'key2':['one','two','one','two','one'],
    'data1':np.random.randn(5),
    'data2':np.random.randn(5)
})
df
```

```python
grouped=df['data1'].groupby(df['key1'])
grouped
```

```python
grouped.mean()
```

```python
means=df['data1'].groupby([df['key1'],df['key2']]).mean()
means
```

```python
means.unstack()
```

```python
for name,group in df.groupby('key1'):
    print(name)
    print(group)
```

```python
for group in df.groupby(['key1','key2']):
    
    print(group)
```

```python
pieces=list(df.groupby('key1'))
pieces
```
