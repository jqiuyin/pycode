```python
from numpy import *
import operator
from os import listdir
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]#获取数据集的行数
    diffMat = tile(inX, (dataSetSize,1)) - dataSet#未知点与每个已知点的向量相减
    sqDiffMat = diffMat**2#然后平方
    sqDistances = sqDiffMat.sum(axis=1)#然后相加
    distances = sqDistances**0.5#开方，获得距离
    sortedDistIndicies = distances.argsort()  #获得距离列表排序后的索引列表   
    classCount={}          #空的字典
    for i in range(k): #迭代k次
        voteIlabel = labels[sortedDistIndicies[i]] #获得第i个与未知点最近点的标签
        classCount[voteIlabel] = classCount.get(voteIlabel,0) + 1 #对标签计数
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)#排序
   
    return sortedClassCount[0][0] #返回频率最高的标签
def file2matrix(filename):
    fr = open(filename)#打开文件
    arrayOLine = fr.readlines()
    numberOfLines = len(arrayOLine)        #获取文件行数
    returnMat = zeros((numberOfLines,3))    #创建一个样本集规模相等的零矩阵
    classLabelVector = []                       #创建一个空的标记向量
    
    index = 0 #初始化索引
    for line in arrayOLine:     #逐行迭代文件
        line = line.strip() #去除空格
        listFromLine = line.split('\t') #每行以制表符分割
        returnMat[index,:] = listFromLine[0:3] #将每行中的样本集存入对应的零矩阵
        classLabelVector.append(int(listFromLine[-1]))#将每行的标签存入对应的标签向量
        index += 1 #索引加一
    return returnMat,classLabelVector #返回样本集与标签向量
def autoNorm(dataSet):
    minVals = dataSet.min(0)#获取数据集中每列的最小值
    maxVals = dataSet.max(0)#获取最大值
    ranges = maxVals - minVals#获取极差
    normDataSet = zeros(shape(dataSet))#创建一个与数据集规模相等的零矩阵
    m = dataSet.shape[0]#获取数据集的行数
    normDataSet = dataSet - tile(minVals, (m,1))#对每一行减去最小值
    normDataSet = normDataSet/tile(ranges, (m,1))   #除于极差
    return normDataSet, ranges, minVals#返回规范化的数据集，极差，最小值
def datingClassTest():
    hoRatio = 0.10      #拿出10%测试
    datingDataMat,datingLabels = file2matrix('datingTestSet2.txt')       #载入数据
    normMat, ranges, minVals = autoNorm(datingDataMat)#规范化数据
    m = normMat.shape[0]#获取数据集规模
    numTestVecs = int(m*hoRatio)#取出10%
    errorCount = 0.0#错误率归零
    for i in range(numTestVecs):#迭代
        classifierResult = classify0(normMat[i,:],normMat[numTestVecs:m,:],datingLabels[numTestVecs:m],3)#计算分类
        print("the classifier came back with: %d, the real answer is: %d" % (classifierResult, datingLabels[i])) 
        if (classifierResult != datingLabels[i]): errorCount += 1.0#如果错误，错误数加一
    print ("the total error rate is: %f" % (errorCount/float(numTestVecs)))#打印错误率
    print (errorCount)#打印错误数




def classifyPerson():
    resultList=["不喜欢","小魅力","大魅力"]#标签集
    percentTats=float(input('游戏时间？'))#特征输入
    ffmiles=float(input('飞行时间？'))
    iceCream=float(input('冰激凌量?'))
    datingDataMat,datingLabels=file2matrix('datingTestSet2.txt')#训练集载入
    normMat,ranges,minVals=autoNorm(datingDataMat)#规范
    inArr=array([ffmiles,percentTats,iceCream])
    result=classify0((inArr-minVals)/ranges,normMat,datingLabels,3)#计算
    print("预测:",resultList[result-1])


```
